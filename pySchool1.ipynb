```python
#info needed
pass_rate = 0.0

#total schools (count of unique values in column, or just school ids)
total_schools = schools["school"].count()
print("Total Schools: ",total_schools)


#total students (count of unique values in column, or just student ids)
total_students = students["Student ID"].count()
print("Total Students: ",total_students)

#total budget (sum column)
total_budget = schools["budget"].sum()
print("Total Budget: ",total_budget)

#avg math score (avg of column)
avg_math = students["math_score"].mean()
print("Avg Math Score: ",avg_math)

#avg reading score (avg of column)
avg_reading = students["reading_score"].mean()
print("Avg Reading Score: ",avg_reading)


#no of math scores >=70, divide by total students*100
percent_pass_math = students.loc[students["math_score"]>=70, ["math_score"]].count()/total_students*100
print("Percent Passing Math: ",percent_pass_math)

#no of reading scores >=70, divide by total students*100
percent_pass_reading = students.loc[students["reading_score"]>=70, ["reading_score"]].count()/total_students*100
print("Percent Passing Reading: ",percent_pass_reading)
#Whhhhhyyyy is this also producing the var name and dtype in my df???

#overall passing rate (avg of two above percents)
pass_rate = (float(percent_pass_math) + float(percent_pass_reading))/2.0
print("Overall Passing Rate: ",pass_rate)

```
